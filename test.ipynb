```python
import pandas as pd
```

```python
# PREPARING THE FULL BALANCE SHEET:

# Reading the data from the Balance Sheet CSV files:
bs1 = pd.read_csv("BalanceSheet2017-21.csv",index_col="Year")                    # reading data from first file
bs2 = pd.read_csv("BalanceSheet2012-16.csv",index_col="Year")                    # reading data from second file
bs3 = pd.read_csv("BalanceSheet2007-11.csv",index_col="Year")                    # reading data from third file
bs4 = pd.read_csv("BalanceSheet2004-06.csv",index_col="Year")                    # reading data from fourth file

# Merging dataframes to get the final Balance Sheet:
tp = (bs2,bs3,bs4)                                                               # creating tuple of dataframes
bal_sheet = bs1                                                                  # initializing the final dataframe
for dataframe in tp:
    bal_sheet = pd.merge(left=bal_sheet,right=dataframe,                         # merging all dataframes (outer join)
                         on="Year",how="outer")   

bal_sheet = bal_sheet.transpose()                                                # transposing the dataframe
bal_sheet
```

```python
# PREPARING THE FULL INCOME STATEMENT:

# Reading the data from the Income Statement CSV files:
bs1 = pd.read_csv("IncomeStatement2017-21.csv",index_col="Year")                 # reading data from first file
bs2 = pd.read_csv("IncomeStatement2012-16.csv",index_col="Year")                 # reading data from second file
bs3 = pd.read_csv("IncomeStatement2007-11.csv",index_col="Year")                 # reading data from third file
bs4 = pd.read_csv("IncomeStatement2004-06.csv",index_col="Year")                 # reading data from first file

# Merging dataframes to get the final Income Statement:
tp = (bs2,bs3,bs4)                                                               # creating tuple of dataframes
inc_statement = bs1                                                              # initializing the final dataframe
for dataframe in tp:
    inc_statement = pd.merge(left=inc_statement,right=dataframe,                 # merging all dataframes (outer join)
                             on="Year",how="outer")   

inc_statement = inc_statement.transpose()                                        # transposing the dataframe
inc_statement
```

```python
# PREPARING THE FULL CASH FLOW STATEMENT:

# Reading the data from the Cash Flow Statement CSV files:
bs1 = pd.read_csv("CashFlowStatement2019-21.csv",index_col="Year")               # reading data from first file
bs2 = pd.read_csv("CashFlowStatement2014-18.csv",index_col="Year")               # reading data from second file
bs3 = pd.read_csv("CashFlowStatement2009-13.csv",index_col="Year")               # reading data from third file
bs4 = pd.read_csv("CashFlowStatement2004-08.csv",index_col="Year")               # reading data from first file

# Merging dataframes to get the final Cash Flow Statement:
tp = (bs2,bs3,bs4)                                                               # creating tuple of dataframes
cf_statement = bs1                                                               # initializing the final dataframe
for dataframe in tp:
    cf_statement = pd.merge(left=cf_statement,right=dataframe,                   # merging all dataframes (outer join)
                             on="Year",how="outer")   

cf_statement = cf_statement.transpose()                                          # transposing the dataframe
cf_statement
```
